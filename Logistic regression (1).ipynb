```python
X = data.iloc[:, [2, 3]].values# Extracts feature variables (columns at index 2 and 3) and converts them to a NumPy array
y = data.iloc[:, 4].values# Extracts the target variable (column at index 4) and converts it to a NumPy array
```

```python
from sklearn.preprocessing import StandardScaler# Import StandardScaler for feature scaling
sc = StandardScaler()# Create an instance of StandardScaler
X = sc.fit_transform(X)# Standardize the feature variables (X) to have mean = 0 and standard deviation = 
```

```python
# Import the train_test_split function for splitting data
from sklearn.model_selection import train_test_split
# Split the dataset into training (75%) and testing (25%) sets
# random_state=0 ensures reproducibility of results
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression# Import LogisticRegression from scikit-learn
model = LogisticRegression()# Create a logistic regression model
model.fit(X_train, y_train)# Train the model using the training data
```

```python
y_pred = model.predict(X_test)# Predict the target values for the test dataset using the trained model
```
